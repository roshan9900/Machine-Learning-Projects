```python
#checking the unique values in the dataframe
for i in df.columns:
    uni = df[i].nunique()
    print(i,' :',uni)
```

```python
#dropping first two columns, as they are index columns
df.drop(['Unnamed: 0','Index'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#checking the null values
df.isnull().sum()
```

1. There few columns with null values.
2. We either drop them or replace them

We are replace null values with median, because we are doing classification, so no need to check the distribution of the values. We can directly imput missing values with median.

```python
# Replacing missing numerical values with median
for i in df.columns:
    if df[i].dtype == 'int64' or df[i].dtype == 'float64':
        df[i].fillna(df[i].median(),inplace=True)
```

```python
#checking the null values again. only categorical columns are having null values
df.isnull().sum()
```

```python
#checking unique values for categorical columns
for i in df.columns:
    if df[i].dtype =='object':
        uni = df[i].nunique()
        print(i,':',uni)
```

```python
#replacing missing values with mode
for i in df.columns:
    if df[i].dtype=='object':
        df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
#checking null values
for i in df.columns:
    null = df[i].isnull().sum()
    print(i,':',null)
```

```python
#using label encoding for categorical columns
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in df.columns:
    if df[i].dtype=='object':
        df[i] = le.fit_transform(df[i])
```

```python
df.head()
```

```python
#saperating features
x = df.drop('FLAG',axis=1)
y = df['FLAG']
```

```python
#scaling the dataset
from sklearn.preprocessing import StandardScaler
xcols = x.columns
sc = StandardScaler()
x = sc.fit_transform(x)
x = pd.DataFrame(x,columns=xcols)
x.head()
```

```python
#spliting the data for train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test  = train_test_split(x,y,test_size=0.3,random_state=1)
```

Model Building

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(y_test,pred)
```

```python
print(classification_report(y_test, pred))
```

```python
from sklearn.svm import LinearSVC
svm = LinearSVC()
svm.fit(x_train,y_train)
```

```python
pred = svm.predict(x_test)
```

```python
print(classification_report(y_test,pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier()
gb.fit(x_train,y_train)
```

```python
pred = gb.predict(x_test)
```

```python
print(classification_report(y_test,pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dtmodel = dt.fit(x_train, y_train)
```

```python
pred = dtmodel.predict(x_test)
```

```python
print(classification_report(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=42)
rf.fit(x_train,y_train)
```

```python
pred = rf.predict(x_test)
```

```python
print(classification_report(y_test,pred))
```

```python
predtrain = rf.predict(x_train)
accuracy_score(y_train, predtrain)
```

As you can see, the training accuracy of the model is 100% and testing accuracy is 99%.

Hyperparameter Tuning

Our most of the models are performing well, we can do hyperparameter tuning for one model.

```python
from sklearn.model_selection import RandomizedSearchCV
rf = RandomForestClassifier(random_state=42)
parameters = {
    'n_estimators': [50, 100, 200, 400]
}

sear = RandomizedSearchCV(estimator=rf,
                         param_distributions=parameters,
                         cv=5,
                         return_train_score=True)
sear.fit(x, y)
```

```python
sear.best_params_
```

```python
parameters = {
    'criterion': ["gini", "entropy", "log_loss"]
}
    
sear = RandomizedSearchCV(estimator=rf,
                         param_distributions=parameters,
                         cv=5,
                         return_train_score=True)
sear.fit(x, y)
```

```python
sear.best_params_
```

```python
parameters = {
    'max_depth': [1, 2, 3, 4]}
sear = RandomizedSearchCV(estimator=rf,
                         param_distributions=parameters,
                         cv=5,
                         return_train_score=True)
sear.fit(x, y)
sear.best_params_
```

```python
parameters = {
    'min_samples_split': [1, 2, 3, 4]}
sear = RandomizedSearchCV(estimator=rf,
                         param_distributions=parameters,
                         cv=5,
                         return_train_score=True)
sear.fit(x, y)
sear.best_params_
```

```python
parameters = {
    'min_samples_leaf' : [1, 2, 3, 4, 5]}
sear = RandomizedSearchCV(estimator=rf,
                         param_distributions=parameters,
                         cv=5,
                         return_train_score=True)
sear.fit(x, y)
sear.best_params_
```

```python
parameters = {
    'ccp_alpha': [0.001, 0.01, 0.1, 1]}
sear = RandomizedSearchCV(estimator=rf,
                         param_distributions=parameters,
                         cv=5,
                         return_train_score=True)
sear.fit(x, y)
sear.best_params_
```

```python
rf = RandomForestClassifier(n_estimators=50,
                           max_depth=4,
                           min_samples_split=3,
                           min_samples_leaf=5,
                           ccp_alpha=0.001,
                           criterion='entropy')
rfmodel = rf.fit(x_train, y_train)
```

```python
predtest  = rfmodel.predict(x_test)
```

```python
predtrain = rfmodel.predict(x_train)
```

```python
print(f'Accuracy score of traning is: {accuracy_score(predtrain, y_train)}')
print(f'Accuracy score of testing is: {accuracy_score(predtest, y_test)}')
```
